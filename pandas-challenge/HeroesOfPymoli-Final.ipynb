#### Some observable trends from the data analysis below! 
* Most players don't spend more than five dollars per purchase on the game add-ons, the average purchase is three dollars and five cents. 

* There are almost 5 times as many purchase records from male players as there are female players, but on average, female players spend around forty cents more per purchase than their male counterparts. However, given that there's only 81 unique female players recorded and 484 unique male players recorded, this purchase difference may not be statistically significant. 

* Most players are under 30, with 66% of recorded players with a purchase being between the age of 15-24. 

* That chunk of the players drive the average purchase price, with 15-19 year olds making an average purchase of three dollars and eleven cents and 20-24 year olds making an average purchase of three dollars and two cents.

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Beginning with purchases DataFrame, bin data by age group. Use same bins/binlabels variables from earlier section

bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]

binlabels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
#Cut purchasing data into bins, add new column for it.

purchases['age_group'] = pd.cut(purchases['age'], bins, labels = binlabels )

purchases.head()
```

```python
#Calculate and store purchase count by age group, starting with creating groupby object

agePurchase = purchases.groupby('age_group')

agePurchaseCount = agePurchase.count()

agePurchaseCount = agePurchaseCount.drop(['gender', 'item_id', 'age', 'purchase_id', 'item_name', 'price'], axis = 1)

agePurchaseCount
```

```python
#Extract Average purchase amount from groupby table, put into list to add to final summary table

agePurchaseAvg = agePurchase.mean()

agePurchaseAvgList = [i for i in agePurchaseAvg['price']]

agePurchaseAvgListRounded = [round(i, 2) for i in agePurchaseAvgList]

agePurchaseAvgListRounded
```

```python
#Add avg purchase amount to final summary table, format
agePurchaseCount['Average Purchase Price'] = agePurchaseAvgListRounded

agePurchaseCount.columns = ['Purchase Count', 'Average Purchase Price']

agePurchaseCount['Average Purchase Price'] = agePurchaseCount['Average Purchase Price'].map("${:,.2f}".format)

agePurchaseCount
```

```python
#Next up! Extract total purchase value from groupby table, put into list to add to final summary table. 

agePurchaseTotal = agePurchase['price'].sum()

agePurchTotal = [i for i in agePurchaseTotal]

agePurchTotal = [round(i, 2) for i in agePurchTotal]

agePurchTotal
```

```python
#Add purchasing totals to summary dataFrame, format appropriately

agePurchaseCount['Total Purchase Value'] = agePurchTotal

agePurchaseCount['Total Purchase Value'] = agePurchaseCount['Total Purchase Value'].map("${:,.2f}".format)

agePurchaseCount
```

```python
#Last but not least! Calculate the average total purchase per person using unique_sn_gender table to get unique players

unique_age_counts = unique_sn_gender.groupby('age_group')

unique_age_counts = unique_age_counts.count()

unique_age_counts_list = [i for i in unique_age_counts['sn']]


purchPerPerson = [i / j for i, j in zip(agePurchTotal, unique_age_counts_list)]

purchPerPerson = [round(i, 2) for i in purchPerPerson]

purchPerPerson
```

```python
#Add purchases per person to final summary table, format.

agePurchaseCount['Average Total Purchase Per Person'] = purchPerPerson

agePurchaseCount['Average Total Purchase Per Person'] = agePurchaseCount['Average Total Purchase Per Person'].map("${:,.2f}".format)

#Display final summary table re: purchasing behavior for each age group!

agePurchaseCount
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order

```python
purchases.head()
```

```python
#Retrieve item id, item name, and item price columns 

popular_items = purchases[['item_id', 'item_name', 'price']]

popular_items.head()
```

```python
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

popular_items_grouped = popular_items.groupby(['item_id', 'item_name'])

popular_items_grouped.head()
```

```python
#Sort/count purchases for each game
popular_items_count = popular_items.groupby('item_name')['item_id'].transform('count')

popular_items['Purchase Count'] = popular_items_count

popular_items = popular_items.sort_values(by = 'item_id')

popular_items_no_dupes = popular_items.drop_duplicates(subset = 'item_id', keep = 'first', inplace = False )

popular_items_no_dupes
```

```python
#Calculate the total purchasing value for popular games, put it in a list for easier adding to summary df

popular_items_sum = popular_items.groupby(['item_id','item_name']).sum()['price']

popular_items_sum_list = popular_items_sum.to_list()

popular_items_sum_list = [round(i,2) for i in popular_items_sum_list]
```

```python
#Add sum column to original summary df, try not to worry about the warning too much
popular_items_no_dupes['Total Purchase Value'] = popular_items_sum_list

popular_items_no_dupes_sorted = popular_items_no_dupes.sort_values(by = ['Purchase Count'], ascending = False)

popular_items_no_dupes_sorted.head()
```

```python
#Format final summary table of most popular items to have $$ and such

#Formatting
popular_items_no_dupes_sorted['price'] = popular_items_no_dupes_sorted['price'].map("${:,.2f}".format)
popular_items_no_dupes_sorted['Total Purchase Value'] = popular_items_no_dupes_sorted['Total Purchase Value'].map("${:,.2f}".format)

##This is the preview of the summary dataframe showing the most popular items sorted by purchase count.
popular_items_no_dupes_sorted.head()
```
